```python
with SearchEngine() as search:
    zipcode1 = search.by_zipcode(10030)
    zipcode2 = search.by_zipcode(10001)
    zipcode3 = search.by_zipcode(10021)
    zipcode4 = search.by_zipcode(10020)

zipcode_list = [zipcode1, zipcode2, zipcode3, zipcode4]
```

```python
print(zipcode1)
print()
print(zipcode2)
print()
print(zipcode3)
print()
print(zipcode4)
```

```python
def zipcode_attrs(zcode_list):
    for z in zipcode_list:
        print('Zipcode = ', z.zipcode)
        print('City = ', z.major_city)
        print('State Abbreviation = ', z.state_abbr)
        print('Population = ', z.population)
        print()
```

```python
zipcode_attrs(zipcode_list)
```
