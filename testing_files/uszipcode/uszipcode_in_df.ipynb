# Function to generate all zip codes in New York

```python
ny_zipcodes = []
res = search.query(state="NY", ascending=False, returns = 0)
for zipcode in res:
    # print(zipcode.zipcode)
    ny_zipcodes.append(zipcode.zipcode)
set(ny_zipcodes)
```

```python
#ny_zipcodes
```

# Function to generate stats with zipcode

```python
# got this code from DSI-6-ATL
metrics = []
for i in ny_zipcodes:
    row = {}
    zipcode = search.by_zipcode(i)
    zips = zipcode.to_dict()
    row['zipcode'] = zips['zipcode']
    row['median_home_value'] = zips['median_home_value']
    row['population'] = zips['population']
    row['occupied_housing_units'] = zips['occupied_housing_units']
    row['median_household_income'] = zips['median_household_income']
    row['population'] = zips['population']
    metrics.append(row)
```

```python
metrics
```

# Convert to a dataframe

```python
state = pd.DataFrame.from_dict(metrics)
```

```python
state.head()
```

```python
state = state[['zipcode', 'median_home_value', 'median_household_income', 'occupied_housing_units', 'population']]
```

```python
state.head()
```

```python
state.shape
```

```python
state.isnull().sum()
```

```python
# zipcode is an object. Maybe make a float?

#state['zipcode'].value_counts(ascending = True)

state['zipcode'] = state['zipcode'].astype(float)
```

# Repeat process for New York City

```python
# now running function for just new york city
nyc_zipcodes = []
res = search.query(state="NY", city='New York', ascending=False, returns = 0)
for zipcode in res:
    # print(zipcode.zipcode)
    nyc_zipcodes.append(zipcode.zipcode)
set(nyc_zipcodes)
```

```python
# got this code from DSI-6-ATL
metrics = []
for i in nyc_zipcodes:
    row = {}
    zipcode = search.by_zipcode(i)
    zips = zipcode.to_dict()
    row['zipcode'] = zips['zipcode']
    row['median_home_value'] = zips['median_home_value']
    row['population'] = zips['population']
    row['occupied_housing_units'] = zips['occupied_housing_units']
    row['median_household_income'] = zips['median_household_income']
    row['population'] = zips['population']
    metrics.append(row)
```

```python
city = pd.DataFrame.from_dict(metrics)
```

```python
city = city[['zipcode', 'median_home_value', 'median_household_income', 'occupied_housing_units', 'population']]
```
